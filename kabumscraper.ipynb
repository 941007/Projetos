```python
import pandas as pd
import numpy as np
import requests
import re
import math
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time


options = webdriver.ChromeOptions()
options.add_argument('--headless')

# instancia do webdriver
driver = webdriver.Chrome(options=options)

url_nvme = 'https://www.kabum.com.br/hardware/ssd-2-5/ssd-pcie-nvme'
url_proc = 'https://www.kabum.com.br/hardware/processadores'
url_ram = 'https://www.kabum.com.br/hardware/memoria-ram'

sites = [url_nvme, url_proc, url_ram]

# Estrutura de dados para armazenar os resultados
dic_produtos = {'marca': [], 'preco': [], 'segmento':[]}

for site in sites:
    
    driver.get(site)

    # Tempo de espera para a página carregar
    time.sleep(5)

    # Cria instância do soup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Procura a div listingCount e traz seu texto
    num_itens = soup.find('div', id='listingCount').get_text().strip()

    # Extrai informações de num_itens
    index = num_itens.find(' ')
    qtd_itens = num_itens[:index]
    # Divide a qtd de itens pelo número de itens para chegar ao número aprox de páginas
    ult_pag_site = math.ceil(int(qtd_itens) / 20)

    # Imprime o que foi encontrado na div listingCount
    print(f"Encontrados {qtd_itens} em {site}")
    
    for i in range(1, ult_pag_site+1):
        url_pag = f'{site}?page_number={i}&page_size=20&facet_filters=&sort=most_searched'
        driver.get(url_pag)
        time.sleep(5)

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        produtos = soup.find_all('div', class_=re.compile('productCard'))

        # Procura o índice da última '/' da direita para a esquerda
        ult_barra_index = site.rfind('/')

        # Extrai tudo após o último '/'
        segmento = site[ult_barra_index + 1:]

        for produto in produtos:
            marca = produto.find('span', class_=re.compile('nameCard')).get_text().strip()
            preco = produto.find('span', class_=re.compile('priceCard')).get_text().strip()

            dic_produtos['marca'].append(marca)
            dic_produtos['preco'].append(preco)
            dic_produtos['segmento'].append(segmento)


# Fecha a instancia do webdriver
driver.quit()

print(dic_produtos)
```

```python
df = pd.DataFrame(dic_produtos)
df.to_csv('precos_kbm.csv', encoding='utf-8', sep=';')
```
